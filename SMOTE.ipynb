Data cleaning & preprocessing was not needed, will get straight to using SMOTE since the data isn't balanced at all.

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , confusion_matrix 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
df= pd.read_csv('migraine_symptom_classification.csv')
df.shape
```

```python
df.describe()
```

```python
df.head()
```

```python
print(df['Type'].value_counts()) #data imbalance
```

```python
X = df.drop("Type", axis=1)
y = df["Type"]
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)
print(y_resampled)
df_resampled = pd.concat([X_resampled, y_resampled], axis=1)
df_resampled.to_csv("migraine_balanced.csv", index=False)
```

```python
df_resampled.shape
```

```python
df_resampled['Type'].value_counts()
```

```python
df_balanced = (
    df_resampled.groupby("Type", group_keys=True)
    .apply(lambda x: x.sample(n=100, random_state=42))
    .reset_index(drop=True)
)
df_balanced = df_balanced.reset_index(level=0).reset_index(drop=True)

df_balanced.to_csv("migraine_balanced_cleaned2.csv", index=False)
```

```python
df_balanced.describe()
```

```python
df_balanced['Type'].value_counts()
```

```python
x = df.drop('Type' , axis = 1)
y = df['Type']
#data splitting 
X_train , X_test , Y_train , Y_test = train_test_split(
    x, y ,test_size=0.2 , random_state=42
)
```

```python
#building Random forest model
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)
# training the model 
rf_model.fit(X_train,Y_train)
```

```python
#prediction for the testing data
Y_pred = rf_model.predict(X_test)
```

```python
#accuracy evaluation for Random forest 
accuracy = accuracy_score(Y_test,Y_pred) * 100
conf_matrix = confusion_matrix(Y_test, Y_pred)
print(f"Random forest accuracy {accuracy:.4f}")
print("Confusion Matrix\n", conf_matrix)
```

```python
#Standard scalar for KNN model
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)

X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

print(X_train_scaled.head())
```

```python
#KNN model 
k_value = 5
knn_model = KNeighborsClassifier(n_neighbors=k_value)
knn_model.fit(X_train_scaled , Y_train)
```

```python
#prediction and evaluation
Y_pred = knn_model.predict(X_test_scaled)
accuracy = accuracy_score(Y_test,Y_pred) * 100
conf = confusion_matrix(Y_test,Y_pred)
print(f"KNN accuracy {accuracy:.4f}")
print("Confusion Matrix\n", conf_matrix)
```

```python
#Making and training naive bayes model
gnb_model = GaussianNB()
gnb_model.fit(X_train,Y_train)
```

```python
#accuracy and evaluation for naive bayes
Y_pred = gnb_model.predict(X_test)
acuuracy = accuracy_score(Y_test , Y_pred) * 100
conf = confusion_matrix(Y_test,Y_pred)
print(f"Naive Bayes accuracy {accuracy:.4f}")
print("Confusion Matrix\n", conf_matrix)
```

```python
# this is desicion tree
dt_model=DecisionTreeClassifier(
    criterion='gini' ,
    max_depth=5,
    random_state=42
)
dt_model.fit(X_train,Y_train)
```

```python
#evaluation for desicion tree
Y_pred = dt_model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred) * 100
conf_matrix = confusion_matrix(Y_test, Y_pred)
print(f"Acurracy is {acuuracy:.4f}")
print("Confusion matrix\n",conf_matrix)
```
